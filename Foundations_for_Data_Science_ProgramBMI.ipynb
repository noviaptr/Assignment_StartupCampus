## Nomor 2

Menghitung BMI setiap siswa dengan rumus:

  BMI= Berat badan (kg)/(Tinggi Badan (m))2

```python
# Mengkonversi tinggi badan dari cm ke m
tinggi_badan = df_F['Tinggi Badan (cm)'] / 100
#Menghitung BMI
df_F['Berat Badan (kg)'] / (tinggi_badan ** 2)
```

## Nomor 3 - 4
3. Membuat  program untuk menghitung BMI
4. Update dataframe df dengan menambahkan hasil perhitungan soal nomor 2, beri nama kolom BMI.

```python
# Membuat Program BMI
# Membuat function dengan nama "BMI" untuk menghitung BMI pada dataframe
def BMI(berat_badan, tinggi_badan):
    return berat_badan / (tinggi_badan ** 2)

# Update DataFrame dengan menambahkan kolom "BMI"
# Menggunakan function BMI dan menambahkan hasilnya ke dalam kolom 'BMI' dalam dataframe df_F
df_F['BMI'] = BMI(df_F["Berat Badan (kg)"], df_F['Tinggi Badan (cm)'] / 100)

# Menampilkan dataframe dengan kolom BMI
df_F
```

## NOMOR 5


Membuat status BMI untuk setiap siswa dari kategori berikut, lalu menyimpan data statusnya dengan nama kolom Status BMI.

  - Normal: 18.5-24.9

  - Overweight: 25.0-29.9

  - Obesity: 30.0-34.9

  - Underweight: < 18.5






```python
# Membuat Status BMI
def status_bmi(BMI):
    if BMI< 18.5:
        return 'Underweight'
    elif 18.5 <= BMI <= 24.9:
        return 'Normal'
    elif 25.0 <= BMI <= 29.9:
        return 'Overweight'
    elif 30.0 <= BMI <= 34.9:
        return 'Obesity'
    else:
        return 'Obesity class II & III'

# Menambahkan kolom baru ke dalam DataFrame dengan nama Status BMI
df_F['Status BMI'] = df_F['BMI'].apply(status_bmi)
df_F
```
